```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
import time
import pandas as pd
from selenium.webdriver.edge.options import Options
```

```python
options = Options()
options.add_argument("--headless") 
service=Service(exucutable_path=r"C:\Users\DO TUAN MINH\Desktop\edgedriver_win64\msedgedriver.exe")
driver=webdriver.Edge(service=service)
driver.get(r"https://fbref.com/en/squads/add600ae/2024-2025/all_comps/Dortmund-Stats-All-Competitions")
time.sleep(5)
table=driver.find_element(By.ID,"stats_standard_combined")

header_row=table.find_element(By.TAG_NAME, "thead").find_elements(By.TAG_NAME, "th")
header_row=header_row[8:][:-1] #Bỏ các phần như playing time,... từ Gls sau PrgR trở đi phải thêm per 90 minutes nữa
row_texts = [row.text for row in header_row]
df=pd.DataFrame(columns=row_texts)





driver.quit()
```

```python
# df=pd.read_csv(r"C:\Users\DO TUAN MINH\Desktop\ben\Learn\BorrusiaDortmund-player-fatigue-analyzer\players_overal_stats_sofar.csv")
prgr_index = df.columns.get_loc('PrgR')
#  modified_col=df.columns[prgr_index+1:]
df.columns = [col+ "_per_90s" if i>=prgr_index+1  else col for i,col in enumerate(df.columns)]
df.columns
```

```python

service=Service(exucutable_path=r"C:\Users\DO TUAN MINH\Desktop\edgedriver_win64\msedgedriver.exe")
driver=webdriver.Edge(service=service)
driver.get(r"https://fbref.com/en/squads/add600ae/2024-2025/all_comps/Dortmund-Stats-All-Competitions")
time.sleep(5)
table=driver.find_element(By.ID,"stats_standard_combined")
rows=table.find_element(By.TAG_NAME, "tbody").find_elements(By.TAG_NAME, "tr")
list=[]
for row in rows[:30]:
    player_name=row.find_element(By.XPATH, ".//th/a").text
    print(player_name)
    cells=row.find_elements(By.TAG_NAME, "td")[1:]
    nationality_element = row.find_element(By.XPATH, './/td[1]/a/span')
    nationality_element="".join(filter(str.isupper,nationality_element.text))
    cells_text=[cell.text for cell in cells][:-1]
    cells_text.insert(0, player_name)
    cells_text.insert(1, nationality_element)
    list.append(cells_text)

new_datapoint=pd.DataFrame(list,columns=df.columns)
df_updated=pd.concat([df,new_datapoint],ignore_index=True)
print(df_updated)
driver.quit()
```

```python
df_updated.to_csv("players_overal_stats_sofar.csv")
```
